```python
import pandas as pd 
import numpy as np 
```

```python
df_1 =  pd.read_csv("EMD_Reconstructed(i80 0400-0415).csv")
df_2 =  pd.read_csv("EMD_Reconstructed(i80 0500-0515).csv")
df_3 =  pd.read_csv("EMD_Reconstructed(i80 0515-0530).csv")
```

```python
df_1
```

# Adding a new column in all i80 dataset 

```python
num_rows = df_1.shape[0]
df_1["Location"] = ['i80']*num_rows
df_1["Period"] = ['4:00PM - 4:15PM']*num_rows
df_1.head()
```

```python
df_2["Location"] = ['i80']*num_rows
df_2["Period"] = ['5:00PM - 5:15PM']*num_rows
df_2.head()
```

```python
df_3["Location"] = ['i80']*num_rows
df_3["Period"] = ['5:15PM - 5:30PM']*num_rows
df_3.head()
```

```python
df_1['Period'].unique()
```

```python
df_1.to_csv('Updated_i80(04-0415).csv', index = False)
df_2.to_csv('Updated_i80(05-0515).csv', index = False)
```

```python
df_3.to_csv("Updated_i80(0515-0530).csv", index = False)
```

## Combining all updated datasets in one(i80)

```python
# Combine the datasets
combined_i80 = pd.concat([df_1,df_2,df_3], axis=0)

# Reset the index of the combined dataset
combined_i80.reset_index(drop=True, inplace=True)
combined_i80
```

```python
combined_i80.to_csv("Combined_i80.csv", index = False)
```
