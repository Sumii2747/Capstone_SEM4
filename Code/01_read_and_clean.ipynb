#### For this project , we are calculating the inter vehicle distance between two cars so , we removed vehicle who changed the lanes 

```python
# Only Non-Lane changing will be taken and removing all Lane changing Vehicle 

previous_lane_id = None
filtered_data = []

for index, row in df_i.iterrows():
    current_vehicle_id = row['Vehicle_ID']
    current_lane_id = row['Lane_ID']
    
    if previous_lane_id is None or current_lane_id == previous_lane_id:
        filtered_data.append(row)
    
    previous_lane_id = current_lane_id

df_i = pd.DataFrame(filtered_data)
```

```python
## Removing the Zero's from preceding, following , space Headway and Time Headway

df_i = df_i[df_i['Preceding'] != 0]
df_i = df_i[df_i['Following'] != 0]
df_i = df_i[df_i['Space_Headway'] != 0]
df_i = df_i[df_i['Time_Headway'] != 0]
```

```python
# Define the desired lane numbers
desired_lanes = [1, 2, 3]

# Filter the dataset for the desired lanes
df_i = df_i[df_i['Lane_ID'].isin(desired_lanes)]

# Remove cases where there is no preceding vehicle
df_i = df_i[df_i['Preceding'].notnull()]


```
