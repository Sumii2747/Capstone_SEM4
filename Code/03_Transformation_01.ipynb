```python
import numpy as np 
import pandas as pd 
```

```python
df_i = pd.read_csv("Transformation_1.csv")
```

```python
df_i = df_i.drop('Unnamed: 0', axis = 1)
df_i
```

```python
# Convert the input variables from feet to meters in df_i DataFrame
df_i['v_Length'] = df_i['v_Length'] * 0.3048
df_i['Space_Headway'] = df_i['Space_Headway'] * 0.3048
df_i['v_Vel'] = df_i['v_Vel'] * 0.3048
df_i['v_Acc'] = df_i['v_Acc'] * 0.3048
df_i['Local_X'] = df_i['Local_X'] * 0.3048
df_i['Local_Y'] = df_i['Local_Y'] * 0.3048

```

```python
## Merge DataFrame based on Preceding Vehicle ID and Frame ID
df_merged = df_i.merge(df_i, left_on=['Preceding', 'Frame_ID'], right_on=['Vehicle_ID', 'Frame_ID'], suffixes=('', '_preceding'))
```

```python
df_merged.to_csv('merged.csv')
```
