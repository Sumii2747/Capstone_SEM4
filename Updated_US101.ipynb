```python
import pandas as pd 
import numpy as np 
```

```python
df_1 =  pd.read_csv("EMD_Reconstructed(us-101 0750am-0805am).csv")
df_2 =  pd.read_csv("EMD_Reconstructed(us-101 0805am-0820am).csv")
df_3 =  pd.read_csv("EMD_Reconstructed(us-101 0820am-0835am).csv")
```

```python
df_1
```

# Adding two column in us 101 datasets

```python
num_rows = df_1.shape[0]
df_1["Location"] = ['us101']*num_rows
df_1["Period"] = ['7:50AM - 8:05AM']*num_rows
df_1.head()
```

```python
df_2["Location"] = ['us101']*num_rows
df_2["Period"] = ['8:05AM- 8:20AM']*num_rows
df_2.head()
```

```python
df_3["Location"] = ['us101']*num_rows
df_3["Period"] = ['8:20AM- 8:35AM']*num_rows
df_3.head()
```

```python
df_1.to_csv('Updated_us101(0750-0805).csv')
df_2.to_csv('Updated_us101(0805-0820).csv')
df_3.to_csv('Updated_us101(0820-0835).csv')
```

## Combining all updated datasets in one(us-101)

```python
# Combine the datasets
combined_us101 = pd.concat([df_1,df_2,df_3], axis=0)

# Reset the index of the combined dataset
combined_us101.reset_index(drop=True, inplace=True)
combined_us101
```

```python
combined_us101.to_csv("Combined_us101.csv", index = False)
```
